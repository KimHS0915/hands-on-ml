```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
(x_train_full, y_train_full), (x_test, y_test) = keras.datasets.cifar10.load_data()

def split_train_valid(data, split_number=5000):
    return data[split_number:], data[:split_number]

x_train, x_valid = split_train_valid(x_train_full)
y_train, y_valid = split_train_valid(y_train_full)
```

```python
def make_model():
    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=[32, 32, 3]))
    for i in range(20):
        model.add(keras.layers.Dense(100, activation='elu', 
                                     kernel_initializer='he_normal'))
    model.add(keras.layers.Dense(10, activation='softmax'))
    return model
```

```python
lrs = [1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 5e-3, 1e-2, 5e-2]
n_epochs = 250
loss = 'sparse_categorical_crossentropy'
early_stopping_cb = keras.callbacks.EarlyStopping(patience=25)
lr_dict = {}

for lr in lrs:
    print(f'Learning_rate : {lr}')
    model = make_model()
    optimizer = keras.optimizers.Nadam(learning_rate=lr)
    model.compile(loss=loss, optimizer=optimizer, metrics=['acc'])
    model_checkpoint_cb = keras.callbacks.ModelCheckpoint(
        'lr_' + str(lr) + '_cifar10_model.h5', save_best_only=True)
    callbacks = [early_stopping_cb, model_checkpoint_cb]
    history = model.fit(x_train, y_train, epochs=n_epochs, 
                        validation_data=(x_valid, y_valid),
                        callbacks=callbacks)
    lr_dict[str(lr)] = {'val_loss': history.history['val_loss'][-1], 
                        'val_acc': history.history['val_acc'][-1]}
    print()
```

```python
model = keras.models.load_model('lr_5e-05_cifar10_model.h5')
model.evaluate(x_valid, y_valid)
```
